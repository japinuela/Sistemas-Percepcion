## Utilizando Open CV

Lectura de una imagen con *cv2.imread*

```python
import cv2
import matplotlib.pyplot as plt

img_bgr = cv2.imread("./lena_color.tiff", cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

Ejemplo de conversión de color a escala de grises con *cv2.cvtColor*

```python
gris = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(gris, cmap="gray")
plt.axis("off")
plt.show()
```

## Exportando imágenes con Open CV
Utilizaremos la función cv2.imwrite

```python
cv2.imwrite("lena_gray.png", gris)
```

## Integrando OpenCV y Numpy

```python
import numpy as np
a = np.array(gris)
print(type(a))
print("shape: ", a.shape)
print("dtype: ", a.dtype) #uint8 . Cada canal tiene un valor de pixel 0 a 255
print("min ", a.min())
print("max ", a.max())
print("media ", a.mean())
print("std ", a.std())
```

```python
import seaborn as sns
sns.histplot(gris.ravel(), bins=256, binrange=(0,255), stat="density", kde=True, label="Histograma escala de gris")
```

## Ecualización de la imagen en escale de grises
Disponemos de la función *cv2.equalizeHist*

```python
gris_eq = cv2.equalizeHist(gris)
a = np.array(gris_eq)
print(type(a))
print("shape: ", a.shape)
print("dtype: ", a.dtype) #uint8 . Cada canal tiene un valor de pixel 0 a 255
print("min ", a.min())
print("max ", a.max())
print("media ", a.mean())
print("std ", a.std())
```

```python
plt.imshow(gris_eq, cmap="gray")
plt.axis("off")
plt.show()
```

```python
sns.histplot(gris_eq.ravel(), bins=256, binrange=(0,255), stat="density", kde=True, label="Histograma escala de gris")
```
