## Utilizando Open CV

Lectura de una imagen con *cv2.imread*

```python
import cv2
import matplotlib.pyplot as plt

img_bgr = cv2.imread("./lena_color.tiff", cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

Ejemplo de conversin de color a escala de grises con *cv2.cvtColor*

```python
gris = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(gris, cmap="gray")
plt.axis("off")
plt.show()
```

## Exportando imgenes con Open CV
Utilizaremos la funcin cv2.imwrite

```python
cv2.imwrite("lena_gray.png", gris)
```
